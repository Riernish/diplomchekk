```python
import pandas as pd
import os
from datetime import datetime

file_path = 'list_tools_listlevel.txt'
data = pd.read_csv(file_path, delimiter='\t')

data['FIRST_DATE'] = pd.to_datetime(data['FIRST_DATE'])

filtered_data = data[(data['LISTLEVEL'] == 1) & (data['FIRST_DATE'] < datetime(2021, 1, 1))]

directory_path = 'data/'

dataframes = {}

for secid in filtered_data['SECID']:
    filename = f"{secid}-D.txt"
    file_path = os.path.join(directory_path, filename)
    
    if os.path.exists(file_path):
        file_data = pd.read_csv(file_path, delimiter='\t')
        
        
        dataframes[secid] = file_data
        
        
        print(f"Data for {filename}:", file_data.head())  
    else:
        print(f"File not found: {filename}")



```

```python
import pandas as pd
import os
from datetime import datetime


filtered_dataframes = {}

for secid, df in dataframes.items():
    if not pd.api.types.is_datetime64_any_dtype(df['begin']):
        df['begin'] = pd.to_datetime(df['begin'])
    
    filtered_df = df[df['begin'] > datetime(2021, 1, 1)]
    
    filtered_dataframes[secid] = filtered_df

all_start_dates = [(secid, df['begin'].iloc[0]) for secid, df in filtered_dataframes.items() if not df.empty]

most_common_date = pd.Series([date for _, date in all_start_dates]).mode()[0]

different_start_dates_secids = [secid for secid, date in all_start_dates if date != most_common_date]

if different_start_dates_secids:
    print("Files with different start dates:", different_start_dates_secids)
else:
    print("All files start on the same date:", most_common_date)

print("Unique start dates found:", set(date for _, date in all_start_dates))

row_counts = {secid: df.shape[0] for secid, df in filtered_dataframes.items()}
common_row_count = pd.Series(list(row_counts.values())).mode()[0]

different_row_counts = {secid: count for secid, count in row_counts.items() if count != common_row_count}

if different_row_counts:
    print("DataFrames with different row counts:")
    for secid, count in different_row_counts.items():
        print(f"SECID: {secid}, Row Count: {count}")
else:
    print("All DataFrames have the same number of rows.")
```

```python
date_sets = {secid: set(df['begin']) for secid, df in filtered_dataframes.items()}


common_dates = set.intersection(*date_sets.values())


aligned_dataframes = {}
for secid, df in filtered_dataframes.items():

    aligned_df = df[df['begin'].isin(common_dates)]
    aligned_dataframes[secid] = aligned_df

aligned_row_counts = {secid: df.shape[0] for secid, df in aligned_dataframes.items()}
print("Row counts after alignment:")
for secid, count in aligned_row_counts.items():
    print(f"SECID: {secid}, Row Count: {count}")

if len(set(aligned_row_counts.values())) == 1:
    print("Success: All DataFrames now have the same number of rows and aligned dates.")
else:
    print("Error: There's still inconsistency in row counts or dates.")
```

```python
from arch import arch_model
import numpy as np

def forecast_last_close(df):
    if len(df) < 2:
        return None, None  

    df = df.copy()
    df['returns'] = np.log(df['open'] / df['open'].shift(1)) * 100  
    df.dropna(subset=['returns'], inplace=True)

    if df.empty:
        return None, None

    
    model = arch_model(df['returns'].iloc[:-1], vol='Garch', p=1, q=2, mean='Zero', dist='t', rescale=False)

    model_fit = model.fit(disp='off', options={'maxiter': 1000})  

    forecast = model_fit.forecast(horizon=1)
    next_vol = forecast.variance.values[-1, -1]

    last_open = df['open'].iloc[-2]
    predicted_return = np.sqrt(next_vol) * np.random.standard_t(df=model_fit.params['nu'])
    predicted_close = last_open * (1 + predicted_return / 100)

    actual_close = df['close'].iloc[-1]

    return predicted_close, actual_close

for secid, df in aligned_dataframes.items():
    predicted_close, actual_close = forecast_last_close(df)
    if predicted_close is not None:
        print(f"SECID: {secid}, Predicted Close: {predicted_close}, Actual Close: {actual_close}")
```

```python
import numpy as np
import pandas as pd
from arch import arch_model
import statsmodels.api as sm

def forecast_close_prices(dataframes):
    forecast_results = {}
    
    for secid, data in dataframes.items():
        data = data.sort_index()
        
        data['returns'] = np.log(data['close'] / data['close'].shift(1))
        data.dropna(inplace=True) 

        if len(data) < 30:  
            print(f"Not enough data to forecast for {secid}")
            continue

        scaled_returns = data['returns'] * 100

        model = arch_model(scaled_returns[-30:], mean='Zero', vol='Garch', p=1, q=1, rescale=False)
        try:
            model_fit = model.fit(disp='off')
        except Exception as e:
            print(f"Model fitting failed for {secid}: {e}")
            continue

        forecast = model_fit.forecast(horizon=1)
        next_vol = forecast.variance.values[-1, 0]

        last_close = data['close'].iloc[-1]
        last_open = data['open'].iloc[-1]
        predicted_pct_change = np.sqrt(next_vol) * np.random.normal() / 100  # Random shock based on forecasted volatility, scaled back
        predicted_close = last_close * (1 + predicted_pct_change)  # Adjusting the last close by the predicted percentage change
        
        # Store forecast result
        forecast_results[secid] = {
            'Open': last_open,
            'Predicted Close': predicted_close,
            'Actual Close': data['close'].iloc[-1] 
        }

    return forecast_results

forecasted_values = forecast_close_prices(aligned_dataframes)
for secid, values in forecasted_values.items():
    print(f"SECID: {secid}, Open: {values['Open']:.2f}, Predicted Close: {values['Predicted Close']:.2f}, Actual Close: {values['Actual Close']:.2f}")
```

```python
import numpy as np
import pandas as pd
from arch import arch_model
import statsmodels.api as sm

def forecast_and_analyze(dataframes):
    forecast_results = {}
    
    for secid, data in dataframes.items():
        data = data.sort_index()
        
        # Log returns for GARCH model
        data['returns'] = np.log(data['close'] / data['close'].shift(1))
        data.dropna(inplace=True)  

        if len(data) < 60: 
            print(f"Not enough data to forecast for {secid}")
            continue

        scaled_returns = data['returns'] * 100

        model = arch_model(scaled_returns[-30:], mean='Zero', vol='Garch', p=1, q=1, rescale=False)
        try:
            model_fit = model.fit(disp='off')
        except Exception as e:
            print(f"Model fitting failed for {secid}: {e}")
            continue
        
        forecast = model_fit.forecast(horizon=1)
        next_vol = forecast.variance.values[-1, 0]
        
        last_close = data['close'].iloc[-1]
        last_open = data['open'].iloc[-1]
        predicted_pct_change = np.sqrt(next_vol) * np.random.normal() / 100
        predicted_close = last_close * (1 + predicted_pct_change)
        
        X = sm.add_constant(data['open'][-60:])
        Y = data['close'][-60:]
        trend_model = sm.OLS(Y, X)
        trend_results = trend_model.fit()
        
        forecast_results[secid] = {
            'Open': last_open,
            'Predicted Close': predicted_close,
            'Actual Close': data['close'].iloc[-1],
            'Trend':  1 if predicted_close > last_open else -1,
            'Confidence': trend_results.rsquared,
            'Predicted Volatility': next_vol
        }

    return forecast_results

forecasted_values = forecast_and_analyze(aligned_dataframes)
for secid, values in forecasted_values.items():
    print(f"SECID: {secid}, Open: {values['Open']:.2f}, Predicted Close: {values['Predicted Close']:.2f}, Actual Close: {values['Actual Close']:.2f}, Trend: {values['Trend']:.4f}, Confidence: {values['Confidence']:.4f}")
```

```python
def balance_portfolio(forecast_results):
    weights = {}

    total_confidence = sum(values['Confidence'] for secid, values in forecast_results.items() if values['Trend'] > 0)
    
    if total_confidence == 0:
        print("No positive trends detected; all weights are set to zero.")
        for secid in forecast_results:
            weights[secid] = 0
        return weights
    
    # Распределяем веса на основе уверенности, если тренд положительный
    for secid, values in forecast_results.items():
        if values['Trend'] > 0:
            weights[secid] = values['Confidence'] / total_confidence
        else:
            weights[secid] = 0  # Устанавливаем вес в ноль, если тренд отрицательный
    
    return weights

forecasted_values = forecast_and_analyze(aligned_dataframes)
portfolio_weights = balance_portfolio(forecasted_values)

for secid, weight in portfolio_weights.items():
    print(f"SECID: {secid}, Weight: {weight:.4f}")
```

```python
#Пробую скользящее окно.

import vectorbt as vbt
import pandas as pd
from arch import arch_model
import numpy as np

def dynamic_scale(data):
    max_val = data['close'].max()
    if max_val > 1000:
        return 0.1 * (1000 / max_val)  # Дополнительное жёсткое масштабирование
    elif max_val < 1:
        return 1 / max_val
    else:
        return 0.1  # Жёстко масштабируем все данные на 0.1

# Определяем размеры окон, которые будем тестировать
window_sizes = [30, 60, 90, 120, 40, 50, 70, 80]

# Результаты будем хранить в списке
results = []

data = df_copy['PIKK'][['begin', 'open', 'close']]
scale_factor = dynamic_scale(data)
data = data.dropna(subset=['close', 'open'])
data.loc[:, 'close_scaled'] = data['close'] * scale_factor

# Модифицированная функция для GARCH с использованием скользящего окна
def garch_predict(data, window_size):
    predictions = []
    for i in range(window_size, len(data)):
        train_data = data.iloc[i-window_size:i]  # Используем последние window_size дней
        model = arch_model(train_data['close_scaled'], vol='Garch', p=1, q=1, rescale=False)
        model_fitted = model.fit(disp='off')
        forecast = model_fitted.forecast(horizon=1)
        predicted_close_scaled = forecast.mean.iloc[-1, 0]
        predicted_close = predicted_close_scaled / scale_factor
        predictions.append(predicted_close)
    return np.array([np.nan] * window_size + predictions)  # Добавляем NaN для первых дней в окне

# Цикл по разным размерам окна
for window_size in window_sizes:
    # Прогнозируем значения с помощью GARCH
    data['Predicted_Close'] = garch_predict(data, window_size)

    # Генерация торговых сигналов
    data.loc[:, 'Signal'] = np.where(data['Predicted_Close'] > data['open'], 1, 0) # 1 - сигнал на покупку
    data.loc[:, 'Exit_Signal'] = data['Signal'].shift(1) # Смещение на один день
    data['Exit_Signal'].iloc[0] = 0  # Первый день без сигнала на продажу

    entries = data['Signal'] == 1
    exits = data['Exit_Signal'] == 1

    price = data['close']
    pf = vbt.Portfolio.from_signals(
        close=price,
        entries=entries,
        exits=exits,
        init_cash=10000,
        freq='1D',
        sl_stop=0.10  # Стоп-лосс на уровне 10% от цены входа
    )

    # Сохраняем результаты для текущего размера окна
    stats = pf.stats()
    results.append((window_size, stats['Total Return [%]'], stats['Sharpe Ratio']))

# Выводим результаты для каждого размера окна
for window_size, total_return, sharpe_ratio in results:
    print(f"Window Size: {window_size}, Total Return: {total_return:.2f}%, Sharpe Ratio: {sharpe_ratio:.2f}")
```
